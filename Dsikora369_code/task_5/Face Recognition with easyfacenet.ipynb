Importing facenet and downloading 3 images: the first one has to be the face that we want to recognize, the second one has to be the wrong one, the last one we want to test.

```
from easyfacenet.simple import facenet
from google.colab import files
files.upload()
images = ['image1.jpg', 'image2.jpg', 'image3.jpg']
```

Using alignment and embedding.

```
aligned = facenet.align_face(images)
embeddings = facenet.embedding(aligned)
```

Creating comparison.

```
comparisons = facenet.compare(aligned)
```

**Comparison**

If you want to get comparison between images, you should get result from comparisons. For example, to compare image 1 to image 2 use:

```
print("Is image 1 and 2 similar? ", bool(comparisons[0][1]))
```

To compare image 1 to image 3 use:

```
print("Is image 1 and 3 similar? ", bool(comparisons[0][2]))
```
